```python
appended_data = []
for x in list(range(0, 200)):
    if x < 10:
        filename = "s3://cads-emory-data/data/parquet/ddc_listings_parq/part-0000" + str(x) + "-e23b3e0d-d711-418e-8461-73d26117709a-c000.snappy.parquet"
        print(filename)
    elif x<100:
        filename = "s3://cads-emory-data/data/parquet/ddc_listings_parq/part-000" + str(x) + "-e23b3e0d-d711-418e-8461-73d26117709a-c000.snappy.parquet"
        print(filename)
    else:
        filename = "s3://cads-emory-data/data/parquet/ddc_listings_parq/part-00" + str(x) + "-e23b3e0d-d711-418e-8461-73d26117709a-c000.snappy.parquet"
        print(filename)
    data = pd.read_parquet(filename)
    data = data[data.groupby('vin')['last_seen_datetime'].transform('max') == data['last_seen_datetime']]
    df1 = data[data['first_seen_datetime']>='2019-08-01']
    df1 = df1[df1['last_seen_datetime']<'2019-11-01']
    print(df1.shape)
    # store DataFrame in list
    appended_data.append(df1)
# see pd.concat documentation for more info
ddc_listings = pd.concat(appended_data)
```

```python
ddc_listings.shape
```

```python
ddc_listings = ddc_listings[ddc_listings.groupby('vin')['last_seen_datetime'].transform('max') == ddc_listings['last_seen_datetime']]
```
