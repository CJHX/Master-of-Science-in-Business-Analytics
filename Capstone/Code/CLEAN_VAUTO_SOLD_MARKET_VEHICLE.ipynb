```python
VASCVIN = VASC['vin']
VASNC[VASNC['vin'].isin(VASCVIN)].shape
```

```python
VASC['Source'] = 'VASC'
VASNC['Source'] = 'VASNC'
```

```python
Combined = pd.concat([VASC, VASNC])
```

```python
Combined = Combined[Combined.vin.str.len() == 17] 
```

```python
Combined['postal_code'] = Combined['postal_code'].astype('str').str.strip()
Combined.loc[(Combined.postal_code.str.len() != 5),'postal_code'] = np.nan
```

```python
Combined['date'] = pd.to_datetime(Combined.EFFECTIVE_START_TS, format='%Y-%m-%d %H:%M:%S.%f')
```

```python
Combined['Key'] = Combined.index
```

```python
Combined['Trim'] = Combined['series'].fillna('').str.cat(Combined['series_detail'].fillna(''),sep=" ")
```

```python
Concise = Combined[['Key','Source','vin','model_year','make','model','Trim','odometer','date']]
```

```python
Concise.head()
```

```python
Concise = Concise.rename(columns={'vin':"Vin",'model_year':"Year",'make':"Make",'model':"Model",'odometer':"Mileage",'date':"Date"})
```

```python
Concise = Concise.replace(r'^\s*$', np.nan, regex=True)
```
