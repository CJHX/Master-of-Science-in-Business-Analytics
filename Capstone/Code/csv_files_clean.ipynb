```python
client_master = pd.read_csv("client_master.csv")
```

```python
NGC_match = client_master[client_master['accttype']=='NGC']
NGC_match = NGC_match[NGC_match['acctsubtype']=='DEALER']
```

```python
NGC_match.head()
```

```python
NGC_match.shape
```

```python
NGC_match = NGC_match.loc[:,['acctnum','caid']]
```

```python
NGC['business_number'] = NGC['business_number'].astype(str)
```

```python
NGC_match = pd.merge(NGC, NGC_match.drop_duplicates(subset=['acctnum']), left_on='business_number', right_on='acctnum',how='left')
```

```python
NGC_match.head()
```

```python
pd.isnull(NGC_match).sum()
```

```python
NGC_match = NGC_match[NGC_match.caid.notnull()]
```

```python
NGC_match.columns
```

```python
cols = ['caid','unit_vin','unit_purchase_price', 'FlooredAmount']
```

```python
NGC_match = NGC_match[cols]
```

```python
NGC_match = NGC_match.rename(columns = {'unit_vin':'Vin'})
NGC_match = NGC_match.rename(columns = {'unit_purchase_price':'NGC_purchase_price'})
NGC_match = NGC_match.rename(columns = {'FlooredAmount':'NGC_floored_amount'})
```
