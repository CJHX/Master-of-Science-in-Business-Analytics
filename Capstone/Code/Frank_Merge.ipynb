```python
Wholesale_cleaned = pd.DataFrame()
```

```python
Wholesale_cleaned['Make'] = Wholesale['MAKE_DESC']
```

```python
Wholesale_cleaned['Model'] = Wholesale['MODEL_DESC']
```

```python
Wholesale_cleaned['Trim'] = Wholesale['TRIM_DESC']
```

```python
Wholesale_cleaned['Mileage'] = Wholesale['ODOMETER']
```

```python
Wholesale_cleaned['Date'] = pd.to_datetime(Wholesale['SOLD_DT_KEY'], format='%Y%m%d')
```
