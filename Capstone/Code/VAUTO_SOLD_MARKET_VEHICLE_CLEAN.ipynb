```python
appended_data = []
for x in list(range(0, 86)):
    y = str(100000 + x)[-5:]
    print(y)
    filename = "s3://cads-emory-data/data/parquet/vauto/vauto_sold_market_vehicle_clean/part-" + y + "-380a2618-1328-4b71-beaa-deef4dd1c8bf-c000.snappy.parquet"
    data = pd.read_parquet(filename)
    print(data.shape)
    # store DataFrame in list
    appended_data.append(data)
# see pd.concat documentation for more info
vauto_atc = pd.concat(appended_data)
```
