```python
appended_data = []
for x in list(range(0, 300)):
    y = str(100000 + x)[-5:]
    print(y)
    filename = "s3://cads-emory-data/data/parquet/vauto/vauto_recent_market_data_new_clean/part-" + y + "-145f3747-45c4-4f54-806b-ab69df87e5ec-c000.snappy.parquet"
    data = pd.read_parquet(filename)
    print(data.shape)
    # store DataFrame in list
    appended_data.append(data)
# see pd.concat documentation for more info
vauto_atc = pd.concat(appended_data)
```
