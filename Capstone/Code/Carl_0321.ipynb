```python
import pandas as pd
import numpy as np
```

```python
VASC = pd.read_csv('vauto_sold_market_vehicle_clean.csv')
VASNC = pd.read_csv('vauto_sold_market_vehicle_new_clean.csv')
```

```python
VASC['Source'] = 'VASC'
VASNC['Source'] = 'VASNC'
```

```python
Combined = pd.concat([VASC, VASNC])
Combined = Combined[Combined.vin.str.len() == 17] 
Combined['postal_code'] = Combined['postal_code'].astype('str').str.strip()
Combined.loc[(Combined.postal_code.str.len() != 5),'postal_code'] = np.nan
```

```python
del VASC
del VASNC
```

```python
Combined['date'] = pd.to_datetime(Combined.EFFECTIVE_START_TS, format='%Y-%m-%d %H:%M:%S.%f')
Combined['Key'] = Combined.index
Combined['Trim'] = Combined['series'].fillna('').str.cat(Combined['series_detail'].fillna(''),sep=" ")
Combined = Combined.replace(r'^\s*$', np.nan, regex=True)
```

```python
Combined.head()
```

```python
Combined.to_csv("Complete_VAUTO_SOLD_MARKET_VEHICLE.csv",index=False)
```

```python
import pandas as pd
import numpy as np
```

```python
VAS = pd.read_csv("VMA_VAUTO_SOLD.csv")
```

```python
VAS.head()
```

```python
print(VAS.shape)
VAS.columns
```

```python
xref = pd.read_csv('vauto_atc_customer_xref.csv')
xref = xref.iloc[:,[0,1,4]]
xref['atc_customer_id'] = xref['atc_customer_id'].astype(str)
```

```python
xref.head()
```

```python
client_master = pd.read_csv("client_master.csv")
```

```python
client_master.head()
```

```python
match = client_master[client_master['accttype']=='SVOC']
match = match[match['acctsubtype']=='PRIMARY']
match = match.iloc[:,[0,3]]
```

```python
match.head()
```

```python
connect = pd.merge(xref,match,how='left', left_on='atc_customer_id', right_on='acctnum')
connect['caid'] = connect.apply(lambda row: 'ATC'+str(row.atc_customer_id) if pd.isnull(row.caid) else row.caid,axis=1)
connect = connect.iloc[:,[0,1,4]]
connect.head()
```

```python
add2 = pd.merge(VAS,connect,how='left', left_on='seller_name', right_on='vauto_seller_name')
```

```python
add2.head()
```

```python
add2['caid'] = add2.apply(lambda row: 'POT'+ str(row.seller_name) if pd.isnull(row.caid) else row.caid,axis=1)
```

```python
print(VAS.shape)
print(add2.shape)
add2.columns
```

```python
add2.head()
```

```python
add2.isna().sum()
```

```python
add2.to_csv('0322_Carl_merge.csv',index=False)
```
