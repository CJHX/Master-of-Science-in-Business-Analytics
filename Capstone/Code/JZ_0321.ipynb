```python
add = pd.read_csv('0321JZ_merge.csv')
```

```python
add.head()
```

```python
add['price'] = add.apply(lambda row: row.list_price if pd.notnull(row.list_price) 
                         else (row.MIN_LIST_PRICE if pd.notnull(row.MIN_LIST_PRICE) 
                               else (row.FIRST_LIST_PRICE if pd.notnull(row.FIRST_LIST_PRICE) 
                                     else np.nan)),axis=1)
```

```python
add1 = add.loc[:,['Key','Source','Vin2','seller_name','price']]
```

```python
add1.head()
```

```python
add1.shape[0]
```

```python
xref = pd.read_csv('vauto_atc_customer_xref.csv')
xref = xref.iloc[:,[0,1,4]]
xref['atc_customer_id'] = xref['atc_customer_id'].astype(str)
```

```python
xref.head()
```

```python
client_master = pd.read_csv("client_master.csv")
```

```python
client_master.head()
```

```python
match = client_master[client_master['accttype']=='SVOC']
match = match[match['acctsubtype']=='PRIMARY']
match = match.iloc[:,[0,3]]
```

```python
match.head()
```

```python
connect = pd.merge(xref,match,how='left', left_on='atc_customer_id', right_on='acctnum')
```

```python
connect['caid'] = connect.apply(lambda row: 'ATC'+str(row.atc_customer_id) if pd.isnull(row.caid) else row.caid,axis=1)
```

```python
connect = connect.iloc[:,[0,1,4]]
```

```python
connect.head()
```

```python
add2 = pd.merge(add1,connect,how='left', left_on='seller_name', right_on='vauto_seller_name')
```

```python
add2.head()
```

```python
add2['dup'] = add2.Key.astype(str) + add2.Source
```

```python
add3 = add2.drop_duplicates(subset=['dup'])
```

```python
add3.shape[0]
```

```python
add3.head()
```

```python
add3['caid'] = add3.apply(lambda row: 'POT'+ str(row.seller_name) if pd.isnull(row.caid) else row.caid,axis=1)
```

```python
add4 = add3.iloc[:,[4,7]]
```

```python
add4.head()
```

```python
add4.shape[0]
```

```python
result = pd.concat([recent.reset_index(),add4.reset_index()], axis=1)
```
