```python
appended_data = []
for x in list(range(0, 62)):
    if x < 10:
        filename = "s3://cads-emory-data/data/parquet/atc_listings_parq/part-0000" + str(x) + "-5d633ec2-9438-44d2-be57-0c1a96522642-c000.snappy.parquet"
        print(filename)
    else:
        filename = "s3://cads-emory-data/data/parquet/atc_listings_parq/part-000" + str(x) + "-5d633ec2-9438-44d2-be57-0c1a96522642-c000.snappy.parquet"
        print(filename)
    data = pd.read_parquet(filename)
    df1 = data[data['SOURCE_CREATED_DATETIME']>='2019-08-01']
    df1 = df1[df1['SOURCE_CREATED_DATETIME']<'2019-11-01']
    print(df1.shape)
    # store DataFrame in list
    appended_data.append(df1)
# see pd.concat documentation for more info
atc_listings = pd.concat(appended_data)
```
